```
import pandas as pd
```

```
#loading results of YOLOv5 Detection
results = pd.read_csv('results.txt', header = None)
results
```

```
coor = []
for i in results[0]:
  a = i.split(' ')
  coor.append(a)
```

```
pd.DataFrame(coor)
```

```
df = pd.DataFrame()
df['center_x'] = pd.DataFrame(coor)[1]
df['center_y'] = pd.DataFrame(coor)[2]
df['width'] = pd.DataFrame(coor)[3]
df['height'] = pd.DataFrame(coor)[4]
df['confidence'] = pd.DataFrame(coor)[5]
df['filename'] = pd.DataFrame(coor)[6]
df
```

```
dict(df.iloc[0])
```

```
def del_element(dict):
  del dict['filename']
  return dict
```

```
relative_coordinates = pd.Series(dict(del_element(df.iloc[i])) for i in range(len(df)))
relative_coordinates
```

```
relative_coordinates[0]
```

```
df['relative_coordinates'] = relative_coordinates
df.drop(['center_x', 'center_y', 'width', 'height', 'confidence'], inplace=True, axis=1)
```

```
df
```

```
data=df
```

```
def details(x,flag_2):
        wid=float(x['width'])/2.0
        hi=float(x['height'])/2.0
        if flag_2=='y2':
            return float(x['center_y'])+hi
        if flag_2=='x2':
            return float(x['center_x'])+wid
        if flag_2=='x1':
            return float(x['center_x'])-wid
        if flag_2=='y1':
            return float(x['center_y'])-hi
data['x1']=data['relative_coordinates'].transform(lambda f:details(f,'x1'))    
data['y1']=data['relative_coordinates'].transform(lambda f:details(f,'y1'))  
data['x2']=data['relative_coordinates'].transform(lambda f:details(f,'x2'))  
data['y2']=data['relative_coordinates'].transform(lambda f:details(f,'y2'))
```

```
data
```

```
count = 0
for i in df['filename']:
  data['filename'].iloc[count] = i.replace("txt", "jpg")
  count += 1
```

```
from PIL import Image
count=0
for x,y in data.iterrows():
  img = Image.open('/content/anpr_dataset/valid/images/'+str(y['filename']))
  dim=img.size
  w,h=dim[0],dim[1]
  x1=y['x1']*w
  x2=y['x2']*w
  y1=y['y1']*h
  y2=y['y2']*h
  area=(x1,y1,x2,y2)
  print(count)
  cropped_img = img.crop(area)
  cropped_img=cropped_img.convert('RGB')
  cropped_img.save("anpr_dataset/valid_cropped/"+str(x)+'.jpg')
  count+=1
```
